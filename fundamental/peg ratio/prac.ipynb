```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stockslist = pd.read_csv('../../esg_nse_yahoo_final.csv')
```

```python
stockslist.head()
```

```python
dic = {}
issues = []
for ticker in stockslist['ticker']:
  try:
    sym = yf.Ticker(f'{ticker}.NS')
    eps = sym.financials['2020-03-31']['Basic EPS']
    hist = pd.read_csv(f'../../data/{ticker}.csv')
    hist['Date'] = pd.to_datetime(hist['Date'])
    price = hist[hist['Date'] == '2020-03-31']['Close'].values[0]
    pe = price/eps
    dic[ticker] = pe
    print(ticker, pe)
  except:
    issues.append(ticker)
    print(ticker, 'Issue')
```

```python
df = pd.DataFrame.from_dict(dic, orient='index')
```

```python
df.columns = ['PE']
```

```python
df.head()
```

```python
df.to_csv('pe.csv')
```

```python
df = pd.read_csv('pe.csv')
df.head()
```

```python
# Sort the DataFrame by the "pe" column in ascending order
df_sorted = df.sort_values(by="PE")

# Filter out rows with negative PE values
df_sorted = df_sorted[df_sorted["PE"] >= 0]

# Calculate the number of rows to skip (5% of the total rows)
skip_rows = int(0.01 * len(df_sorted))

# Slice the DataFrame to exclude the first 5%
df_filtered = df_sorted.iloc[skip_rows:]

# Get the top 40 stocks with the smallest "pe" values
top_40_stocks = df_filtered.head(40)

# Reset the index if needed
top_40_stocks.reset_index(drop=True, inplace=True)
```

```python
top_40_stocks.head()
```

```python
top_40_stocks.columns = ['Ticker', 'PE']
```

```python
pd.DataFrame(top_40_stocks['Ticker']).to_csv('top_40_stocks.csv', index=False)
```
